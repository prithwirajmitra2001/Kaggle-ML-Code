```python
import pandas as pd
import numpy as np
import scipy.optimize as op
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Cross Validation (K-fold)

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
from sklearn. ensemble import BaggingRegressor, AdaBoostRegressor, VotingRegressor
```

```python
train = pd.read_csv (r"C:\Users\admin\Desktop\Datasets\House Prices Advanced Regression Techniques Kaggle\train_clean.csv")
test = pd.read_csv (r"C:\Users\admin\Desktop\Datasets\House Prices Advanced Regression Techniques Kaggle\test_clean.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
print (train.shape)
print (test.shape)
```

```python
x = train.iloc[:,0:73]
y = train.iloc[:,-1]
```

```python
x_train , x_test , y_train , y_test = train_test_split (x,y,test_size=0.33,random_state=5)
```

```python
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
# Multiple Linear Regression
```

```python
# Random Search

hyperparameter_grid={
    'fit_intercept'      : [True , False , None],
    'normalize'          : [True , False , None],
    'n_jobs'             : [-1]
}
```

```python
regressor = LinearRegression()
random_cv = RandomizedSearchCV(estimator=regressor , param_distributions=hyperparameter_grid , cv=5 , random_state=42 ,
                               scoring = 'neg_mean_absolute_error',n_jobs = -1, verbose = 5 , return_train_score = True )
random_cv.fit(x_train , y_train)
```

```python
random_cv.best_estimator_
```

```python
regressor = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=-1, normalize=False)
regressor.fit(x_train,y_train)
prediction1 = regressor.predict(test)
```

```python
score_train = cross_val_score(regressor, x_train,y_train, cv=k_fold, n_jobs=-1)
print (round(np.mean(score_train)*100,2))
score_test = cross_val_score(regressor, x_test,y_test, cv=k_fold, n_jobs=-1)
print (round(np.mean(score_test)*100,2))
```
